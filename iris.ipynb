No missing values, but we can just do it for practice

```python
num_col=df.select_dtypes(include=['int64','float64']).columns
cat_col=df.select_dtypes(include=['object','category']).columns
```

```python
num_missing=[col for col in num_col if df[col].isnull().sum()>0]
cat_missing=[col for col in cat_col if df[col].isnull().sum()>0]
```

```python
if len(num_missing) > 0:
    df[num_missing].describe()
else:
    print("No numeric columns with missing values")
```

```python
if len(cat_missing) > 0:
    print(df[cat_missing].isnull().sum())
else:
    print("No categorical columns with missing values")
```

Identify if there is Outliers

```python
import matplotlib.pyplot as plt

for col in num_col:
    plt.hist(df[col].dropna(),bins=30)
    plt.title(col)
    plt.show()
```

```python
for col in num_col:
    print(col)
    print("Skew:", df[col].skew())
    print("Min:", df[col].min(), "Max:", df[col].max())
    print("-" * 30)
```
