Step 3 — Handle Outliers

```python
#Focus on numeric columns
df.select_dtypes(include='number').columns
```

```python
#Detect outliers using IQR (for total_bill)
Q1=df['total_bill'].quantile(0.25)
Q3=df['total_bill'].quantile(0.75)
IQR=Q3-Q1

lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
```

```python
#Cap outliers
df['total_bill']=df['total_bill'].clip(lower,upper)
```

Step 4 — Feature Scaling

```python
num_cols = ['total_bill', 'tip', 'size']
```

```python
from sklearn.preprocessing import StandardScaler


scaler=StandardScaler()
df[num_cols]=scaler.fit_transform(df[num_cols])
```

Step 5 — Feature Engineering

```python
#Create a new feature: Tip Percentage

df['tip_percent']=df['tip']/df['total_bill']
```

```python
df
```

```python
#Encode categorical variables

df=pd.get_dummies(df,columns=['sex','smoker','day','time'],drop_first=True)
```

Step 6 — Feature Selection

```python
#Correlation with target (tip)
corr=df.corr()['tip'].sort_values(ascending=False)
corr
```

```python
#Select useful features
selected=corr[abs(corr)>0.1].index
df=df[selected]
```
